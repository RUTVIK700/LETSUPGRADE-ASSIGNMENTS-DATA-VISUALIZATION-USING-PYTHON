```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#QUESTION-1
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b,'go--',linewidth=3,markersize=6)
plt.title('LINE PLOT BETWEEN a AND b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#QUESTION-2:-Plot a line plot showing the sales trend in company 1 and 2:
days = [1,2,3,4,5,6,7] 
sales_1 = [160,150,140,145,175,165,180] 
sales_2 = [70,90,160,150,140,145,175]  
plt.title("Sales trend in company 1 and 2")
plt.plot(days, sales_1,'--',linewidth=2,markersize=5,color='g',marker ='o',label='sales_1')
plt.plot(days, sales_2,'--',linewidth=2,markersize=5,  color='b' , marker ='o',label='sales_2')
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend(loc='lower center')
plt.show()
```
